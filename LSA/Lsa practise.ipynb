```python
df= pd.read_csv(r"C:\Users\Sahil\Videos\test.csv")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD
from sklearn.metrics.pairwise import cosine_similarity
```

```python
vectorizer=TfidfVectorizer(stop_words='english')
fitting=vectorizer.fit_transform(df)
```

```python
Latent_Semantic_Analysis=TruncatedSVD(n_components=3,random_state=42)
box=Latent_Semantic_Analysis.fit_transform(fitting)
```

```python
topics=vectorizer.get_feature_names_out()
for i, comp in enumerate(Latent_Semantic_Analysis.components_):
    print(f"\nüóìÔ∏èTopic {i+1}")
    topic_weights= zip(topics,comp)
    topic=sorted(topic_weights,key=lambda x:x[1],reverse=True)
    for word,weight in topic[:10]:
        print(word)
```
