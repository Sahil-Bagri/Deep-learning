```python
import pandas as pd
```

```python
df=pd.read_csv(r"C:\Users\Sahil\OneDrive\Documents\test.csv")
```

```python
df.head()
```

```python
df.drop(["Title","Class Index"],axis=1)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_df=0.9,min_df=2,stop_words='english')
x=cv.fit_transform(df['Description'])
```

```python
x
```

```python
from sklearn.decomposition import LatentDirichletAllocation
```

```python
LDA = LatentDirichletAllocation(n_components=4,random_state=42)
LDA.fit(x)
```

```python
import random
```

```python
for i in range(20):
  y= random.randint(0,10529)
  print(cv.get_feature_names_out()[y])
```

```python
len(LDA.components_)
```

```python
LDA.components_
```

```python
len(LDA.components_[0])
```

```python
first_topic = LDA.components_[0]
```

```python
first_topic.argsort()
```

```python
first_topic.argsort()[-20:]
```

```python
top_words = first_topic.argsort()[-20:]
```

```python
for z in top_words:
  print(cv.get_feature_names_out()[z])
```

```python
for z,topic in enumerate(LDA.components_):
  print(f'the top4 words for topic {z}')
  print([cv.get_feature_names_out()[a] for a in topic.argsort()[-20:]])
  print('\n')
```
